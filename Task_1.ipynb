# Task 1

## Import required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv(
    "Population_task_1.csv",
    skiprows=4
)
```

## View first 5 rows

```python
df.head()
```

## Check Column

```python
df.columns
```

## Drop Unwanted Column

```python
df = df.drop(columns=['Unnamed: 69'])
```

## Verify columns again

```python
df.columns
```

## Select Year for Analysis

```python
year = '1975'
```

## Convert Data Type

```python
df[year] = pd.to_numeric(df[year], errors='coerce')
```

## Handle Missing Values

```python
year_data = df[['Country Name', year]].dropna()
```

## Create Histogram (Continuous Variable)

```python
plt.figure(figsize=(8, 5))
plt.hist(year_data[year], bins=15)
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Population Distribution Across Countries in 1975')
plt.show()
```

## Create Bar Chart (Categorical Variable)

```python
top_10 = year_data.sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_10['Country Name'], top_10[year])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Most Populated Countries in 1975')
plt.xticks(rotation=45)
plt.show()
```
